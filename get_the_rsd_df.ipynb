### Select and extract patients solved confirmed based on solveRD input :


```python
patients_json = os.listdir(path_runsolveRD_CO)

patients_co_files = []
for one_file in patients_json:
    splited = one_file.split('.')[0]
    if splited in patients_c:
        patients_co_files.append(one_file)
        p_index = patients_c.index(splited)
        # verification 
        print("{}\t{}".format(one_file,patients_c[p_index] ))
```

### Keep only the .json.json files :

```python

match_jsonresult = []
for onefile in patients_co_files:
    if ".json.json" in onefile:
        match_jsonresult.append(onefile)
match_jsonresult
```

### Extract the CO files

```python
type_score = 'Phenodigm Similarity'
""" 
Cosine Weighted Similarity 
Cosine Similarity
Bernoulli with grid Resnik (asymmetric) 
Resnik (symmetric)
Jaccard standard
Jaccard weighted
Phenodigm Similarity
"""
```

```python
case_ORPHA = []
for onefile in match_jsonresult:
    try:
        # open one phenopact result per loop
        with open(path_runsolveRD_CO+str(onefile),'r',encoding = 'utf8') as file_phenopacket_result:
            one_phenopacket_result = json.load(file_phenopacket_result,encoding = 'utf8')

            # select only the result from a specific algo (these is 8 algo)
            for key_method in one_phenopacket_result['methods']:
                # go inside the dico which has the key name methods
                if key_method['method'] == type_score:
                    # key_method contains all information of the phenopacket
                    for key_results in key_method['results']:
                        # split part from 'P0000037.json.json' to 'P0000037,
                        split_jsonfile = onefile.split('.')
                        split_jsonfile = split_jsonfile[0]
                        try:
                            case_ORPHA.append((split_jsonfile,key_results['itemid'],float(key_results['score']),key_results['rank']))
                        except TypeError:
                            print("warning Nonetype on {}".format(split_jsonfile))
    except (json.JSONDecodeError) as e:
        print("A2\tError reading the file {}\t{}".format(onefile,e))
        sys.exit(1)
df_case_ORPHA = pd.DataFrame(case_ORPHA, columns=["phenopacket",'ORPHAcode',type_score,'rank'])


df_case_ORPHA
```

```python
df_case_ORPHA.to_excel(PATH_output+"/df_rsd_"+str(type_score)+".xlsx")
```
